## Cleaning columns
Finding non regular rows

Finding categorical rows because we need to make them into integers \\
In this case there are two rows: Winner and title_bout

```python
data.select_dtypes(exclude=["number"])
```

```python
categorical_columns = ["Winner", "title_bout"]
```

```python
for col in categorical_columns:
    data[col] = data[col].astype('category')
```

```python
data = pd.get_dummies(data, drop_first=True)
```

Once more inspecting the data

```python
data.head()
```

## Finding NaN or all zeros columns
There are two columns that have all zeros, B_draw and R_draw. We know that in UFC we can't have draw so we can just drop this two columns

```python
for col in data.columns:
    if(data[col].sum() == 0):
        print(col)
```

```python
data = data.drop(["R_draw", "B_draw"], axis=1)
```
